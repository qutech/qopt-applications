# Saving Data

## Bloch Sphere Plots

```python
# generate an example
import numpy as np
from qsim.matrix import DenseOperator
from qsim.solver_algorithms import SchroedingerSolver


sigma_x = DenseOperator(np.asarray([[0, 1], [1, 0]]))
solver = SchroedingerSolver(
    h_ctrl=[sigma_x],
    h_drift=[0 * sigma_x],
    tau=np.ones(1)
)
solver.set_optimization_parameters(.25 * np.pi * np.ones((1, 1)))
b = solver.plot_bloch_sphere(return_Bloch=True)
```

Option 1: No arguments -> the bloch sphere is saved to the current working
directory with automatically generated name.

```python
b.save()
```

Option 2: Specify the full name including the working directory and format.

```python
import os
current_directory = os.getcwd()
file_name = 'test_bloch.png'
# The join method works independently from the architecture:
name = os.path.join(current_directory, file_name)
b.save(name=name)
```

## Matplotlib Figures

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot([1,2])

fig.savefig('test_plt1.png', bbox_inches='tight')

# or

file_name = 'test_plt2.png'
name = os.path.join(current_directory, file_name)
fig.savefig(name, bbox_inches='tight')
```

## Numpy Arrays

Option 1: Use numpy.save and numpy.load.

```python
test_array = np.arange(4)
np.save('test_array.npy', test_array)
test_array_loaded = np.load('test_array.npy')
test_array_loaded
```

Option 2: Use ndarray.tofile and numpy.fromfile. Remember to specify the data
type in this case!

```python
test_array = np.arange(4)
test_array.tofile('test_array.npy')
test_array_loaded = np.fromfile('test_array.npy', dtype=int)
test_array_loaded
```
